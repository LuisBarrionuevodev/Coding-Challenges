```python
import pandas as pd
tabla_actas= pd.read_excel("./actas.xlsx")
tabla_contestaciones=pd.read_excel("./contestaciones_oficios.xlsx")
tabla_oficios=pd.read_excel("./oficios.xlsx")
tabla_actas
```

```python

actas_merge=tabla_oficios.merge(tabla_contestaciones[["oficio","Tipo","Fecha"]],left_on="oficio",right_on="Oficio",how="left")
actas_merge.to_excel("oficios_resuletos.xlsx",index="false")
tabla_resueltos=pd.read_excel("./oficios_resuletos.xlsx")
```

```python
actas_merge=tabla_actas.merge(tabla_resueltos[["Fecha de asunto","Causante","oficio","Tipo","Fecha","acta"]],left_on="acta",right_on="acta",how="left")

actas_resueltas=actas_merge.to_excel("actas_resueltas.xlsx",index=False)
```

```python
merge_swift= tabla_contestaciones.merge(tabla_oficios[["Asunto","Generado por",	"Fecha de asunto","Causante","oficio","acta"]],left_on="Oficio",right_on="oficio", how="left")
```

```python
merge_swift.to_excel("contestaciones_a.xlsx",index=False)

```

```python
datos_sueltos=merge_swift[merge_swift["Asunto_y"].isna()]
datos_sueltos.to_excel("contetaciones_sueltas.xlsx",index=False)
```
